```python
import scoring_methods as score_meth
import scoring_tester as score
import pandas as pd
import numpy as np
```

```python
scoring_functions = [
    score_meth.structural_similarity,
    score_meth.dist_chebishev, 
    score_meth.dist_euclidean, 
    score_meth.dist_manathan,
    score_meth.normalized_rmse
]
results = {}
score_tester = score.ScoringTester()
```

```python
for function in scoring_functions:
    score_tester.image_matching_func = function
    fun_result = score_tester.aggregated_process()
    if 'population' not in results:
        results['population'] = fun_result['population']
    results['median_' + function.__name__] = fun_result['median']
```

```python
np.save('score_meth_comp_result_dict.npy', results)
```

```python
def color_gradiant(val):
    max = val.max()
    ratio = np.round(255 * val/max)
    return 'color: #' + str(hex(256 - ratio))  + str(hex(ratio)) +'00'
```

```python
# Optionnel pour ne pas tout recalculer
results = np.load('score_meth_comp_result_dict.npy',allow_pickle='TRUE').item()
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 200)
```

```python
df = pd.DataFrame(results).sort_values(by=['population'], ascending=False)
a = df.style.background_gradient('RdYlGn')
a
```

```python
norm_df = df
norm_df['median_dist_chebishev'] = norm_df['median_dist_chebishev'].apply(lambda x: 1-x if x != 0 else 0)
norm_df['median_dist_euclidean'] = norm_df['median_dist_euclidean'].apply(lambda x: (1-(x/np.sqrt(2))) if x != 0 else 0)
norm_df['median_dist_manathan'] = norm_df['median_dist_manathan'].apply(lambda x: (1-(x/10)) if x != 0 else 0)
norm_df['median_normalized_rmse'] = norm_df['median_normalized_rmse'].apply(lambda x: (1-x) if x != 0 else 0)


n_df = norm_df.style.background_gradient('RdYlGn')
n_df
```
