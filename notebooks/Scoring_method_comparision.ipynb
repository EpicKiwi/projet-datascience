```python
import ast
import scoring_methods as score_meth
import scoring_tester as score
import pandas as pd
import numpy as np
```

```python
scoring_functions = [
    score_meth.structural_similarity,
    score_meth.dist_chebishev, 
    score_meth.dist_euclidean, 
    score_meth.dist_manathan,
    score_meth.normalized_rmse,
    score_meth.psnr,
]
results = {}
score_tester = score.ScoringTester()
```

```python
for function in scoring_functions:
    score_tester.image_matching_func = function
    fun_result = score_tester.aggregated_process()
    if 'population' not in results:
        results['population'] = fun_result['population']
    results['mean_' + function.__name__] = fun_result['mean']
```
